```python
import warnings
warnings.filterwarnings('ignore')
# plots
import matplotlib.pyplot as plt
import seaborn as sns

# feature engineering
from src.data.feature_engineering import FeatureEngineering

# helpers
from helpers.config import load_config
config = load_config()

df = FeatureEngineering(config).select_features()

plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),fmt=".2f",annot=True,cmap="coolwarm")
plt.tight_layout()
plt.savefig("images/heatmap-airplane-sales.png")
plt.show()

fig,axs = plt.subplots(2,2,figsize=(10,6))
sns.barplot(data=df,ax=axs[0,0],x='tdrag',y='wtop')
sns.lineplot(data=df,ax=axs[0,1],x='pass',y='0Sale_ID')
sns.kdeplot(data=df,ax=axs[1,0],x='0Sale_ID',hue="tdrag")
sns.countplot(data=df,ax=axs[1,1],x='pass',hue='fixgear')
plt.tight_layout()
plt.show()
```
